```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
N = 1_000_000

# Prior is bimodal
p = np.concatenate(
    [
        np.random.normal(loc=0, scale=1, size=int(N * 0.55)),
        np.random.normal(loc=5, scale=1, size=int(N * 0.45)),
    ]
)

# Posterior is gaussian
q = np.random.normal(1, 1.5, N // 2)

df = pd.concat(
    [pd.DataFrame({"Dist": "p", "Data": p}), pd.DataFrame({"Dist": "q", "Data": q})]
).reset_index(drop=True)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
pal = ["#E71D36", "#2D7DD2"]
sns.kdeplot(
    data=df,
    hue="Dist",
    x="Data",
    shade=True,
    palette=pal,
    legend=False,
    ax=ax,
    bw_adjust=3,
    clip=[-4.5, 9],
    hue_order=['q', 'p']
)
ax.axis("off")
plt.tight_layout()
```
